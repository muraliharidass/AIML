### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
#your code here
# create dataframe with
# release year as index
# genre as column names
# count of releases by genre as row values
# add total as a column

df_yearwise = df_movie.groupby('release year').sum()

# drop unwanted columns
del df_yearwise['movie id']

del df_yearwise['genre_sum']

# add total per year as column
df_yearwise['total_movie_count'] = df_yearwise.sum(axis=1, numeric_only=True)

# df_yearwise.head(10)
```

```python
# df_yearwise.info()
```

```python
#To calculate percentage of movies released for a genre in a year
genre_percent = (df_yearwise.div(df_yearwise['total_movie_count'], axis= 0) * 100)

# To drop total movie count column
del genre_percent['total_movie_count']
```

```python
# To display heatmap of percentage changes of movie genres over the years
ax = sns.heatmap(genre_percent, cmap="YlGnBu")
```

**Insights:**
In the period when movies started receiving ratings, Drama & Horror were the only genres released
Over time, number of movies in other genres have increased
In the whole period between 1922 and 1997, Drama has been a very popular genre
Western films had their peaks in the 1960s
Fantasy and Sci-Fi films have always been less popular


### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here
# a. Count(movie id), average(ratings) group by movie_id
df_movie_smry = df_rating.groupby('movie id',as_index=False)['rating'].count()
df_movie_smry.head()
```

```python
df_movie_smry['avg rating'] = df_rating.groupby('movie id')['rating'].mean()
df_movie_smry.rename(columns = {"rating":"rating count"})
df_movie_smry.head()
```

```python
# df_movie_smry.info()
```

```python
# Merging with df_movie to bring movie title
df_movie_summary = df_movie_smry.merge(df_movie, on=['movie id'])[['movie id','movie title', 'rating','avg rating']]
```

```python
# Final Output - Top 25 movies in terms of average ratings for movies that have been rated more than 100 times
df_movie_summary[df_movie_summary['rating']> 100].sort_values(by='avg rating', ascending=False).head(25)
```

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#your code here
# 1. Merge all the datasets
# Create combined data frame consisting of df_rating, df_user and df_movie
df_combined = pd.merge(pd.merge(df_rating, df_user, how='inner', on='user id'), df_movie, how='inner', on='movie id')
df_combined.head()
```

```python
# find value_counts of gender in df2 that will give the total number of female and male 
# considered for rating
print('Total count of Females and Males who rated the movies')
df_combined.groupby('gender')['user id'].count()
```

```python
# Drama Percentage
df_combined.groupby('gender')['Drama'].sum().div(df_combined.groupby('gender')['user id'].count(), axis=0)*100
```

Out of all the women who rated movies, 42.77% rated Drama whereas only 38.9% men rated Drama

```python
# Sci-Fi Percentage
df_combined.groupby('gender')['Sci-Fi'].sum().div(df_combined.groupby('gender')['user id'].count(), axis=0)*100
```

Out of all the women who rated movies, 10.21% rated Sci-Fi whereas 13.6% men rated Sci-Fi

```python
# Romance Percentage
df_combined.groupby('gender')['Romance'].sum().div(df_combined.groupby('gender')['user id'].count(), axis=0)*100
```

Out of all the women who rated movies, 22.77% rated Romance whereas only 18.32% men rated Romance
